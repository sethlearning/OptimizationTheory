## Практикум 4.1 КОНСТРУИРОВАНИЕ БИНАРНОГО КЛАССИФИКАТОРА

Выполните задание для получения дополнительных баллов (на повышенную оценку).

Необходимо построить и обучить нейросеть, которая на вход берет два числа (x,y) и на выходе дает 0 или 1 по правилу:

если (x<0.5  И y>0.5) или (x>0.5  И y<0.5), то выход = 1 (близок к 1). 

В противном случае должна давать 0.

Вы должны запрограммировать функцию build_nnet(), которая на выходе дает обученную модель (class model из библиотеки Keras).

```python
# Модуль создания и обучения нейросетевого бинарного классификатора
import numpy as np
import random
import tensorflow as tf
from keras.layers import Input, Dense
from keras.models import Model
from keras.optimizers import Adam



def preparedata():
    trainxy = []
    trainz = []
    for i in range(101):
        for j in range(101):
            x = i/100
            y = j/100
            trainxy.append((x, y))
            if (x < 0.5 and y > 0.5) or (x > 0.5 and y < 0.5):
                trainz.append(1)
            else:
                trainz.append(0)

    return np.array(trainxy), np.array(trainz)

def preparedata2(i):
    trainxy = []
    trainz = []
    for _ in range(i):
        xy = np.random.randn(2)
        trainxy.append(xy)
        if (xy[0] < 0.5 and xy[1] > 0.5) or (xy[0] > 0.5 and xy[1] < 0.5):
            trainz.append(1)
        else:
            trainz.append(0)

    return np.array(trainxy), np.array(trainz)

def preparedata3(n):
    trainz = []
    # trainxy = np.random.rand(n, 2)
    trainxy = np.random.randn(n, 2) + 0.5
    for i in range(n):
        if (trainxy[i][0] < 0.5 and trainxy[i][1] > 0.5) or (trainxy[i][0] > 0.5 and trainxy[i][1] < 0.5):
            trainz.append(1)
        else:
            trainz.append(0)
    return np.array(trainxy), np.array(trainz)

def build_nnet():

    model = tf.keras.models.Sequential()

    # конструируем нейросеть
    model.add(Dense(16, activation='sigmoid', input_shape=(2,)))
    model.add(Dense(1, activation='sigmoid'))

    # print(model.summary())
    # подготавливаем данные
    trainxy, trainz = preparedata3(10000)
    testxy, testz = preparedata3(1000)

    # компилируем модель
    model.compile(
        loss = tf.keras.losses.binary_crossentropy,
        # loss = tf.keras.losses.mse,
        optimizer = tf.keras.optimizers.Adam(learning_rate=0.001),
        metrics = ['accuracy']
    )

    # обучаем модель
    batchsize = 40
    epochs = 100
    model.fit(
        trainxy, trainz,
        batch_size = batchsize,
        epochs = epochs,
        verbose = 1,
        validation_data = (testxy, testz)
    )

    return model
```

```python
m = build_nnet()
```

```python
testxy3, testz3 = preparedata3(1000)
loss, accuracy = m.evaluate(testxy3, testz3)
print('Test loss:', loss)
print('Test accuracy:', accuracy)
```
