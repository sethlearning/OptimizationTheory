## Тема 1. Одномерная оптимизация

```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# рассмотрим простую функцию и нарисуем ее график
def f1(x):
    return x**2 - np.sin(x)

x = np.arange(-5, 5, 0.1)
y = (f1(x))

plt.plot(x, y)
```

```python
# возьмем произвольные приближения x1=0 и x2=0.5 и оценим их точность 
x1 = 0
x2 = 1

# для того, чтобы оценить точность по аргументу необходимо знать точное решение!
res = optimize.minimize(f1, x1)
print(res)
x_opt = res.x[0]
```

```python
# а) по аргументу
print(x_opt)
print('погрешность по аргументу для x1=', x1, 'равна', abs(x1-x_opt))
print('погрешность по аргументу для x2=', x2, 'равна', abs(x2-x_opt))
```

```python
# б) по производной
# для этого надо вычислить производную численно или аналитически задать
def df1dx(x):
    return 1*x - np.cos(x)

print('погрешность по производной для x1=', x1, 'равна', abs(df1dx(x1)))
print('погрешность по производной для x2=', x2, 'равна', abs(df1dx(x2)))
```

#### Пример плохо обусловленной задачи

```python
# рассмотрим простую функцию и нарисуем ее график
def f2(x):
    return x**8 - 5*np.sin(x)

x = np.arange(-5, 5, 0.1)
y = (f2(x))

plt.plot(x, y)
```

```python
# рассмотрим те же приближения x1=0 и x2=0.5 и оценим их точность 
x1 = 0
x2 = 1

# для того, чтобы оценить точность по аргументу необходимо знать точное решение!
res2 = optimize.minimize(f2, x1)
print(res2)
x_opt = res2.x[0]
```

```python
# а) по аргументу
print(x_opt)
print('погрешность по аргументу для x1=', x1, 'равна', abs(x1-x_opt))
print('погрешность по аргументу для x2=', x2, 'равна', abs(x2-x_opt))
```

```python
# б) по производной
# для этого надо вычислить производную численно или аналитически задать
def df2dx(x):
    return 8*x**7 - 5*np.cos(x)

print('погрешность по производной для x1=', x1, 'равна', abs(df2dx(x1)))
print('погрешность по производной для x2=', x2, 'равна', abs(df2dx(x2)))
```

#### Пример многоэкстремальной функции

```python
# зададим функцию с несколькими экстремумами
def f(x):
    return x**2 + 8*np.sin(x)

x = np.arange(-7, 7, 0.2)
plt.plot(x, f(x))
```

```python
optimize.minimize(f, 5)
```

```python
optimize.minimize(f, -5)
```
