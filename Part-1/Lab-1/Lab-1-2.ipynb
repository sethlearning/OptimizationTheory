Тип работы:  Индивидуальная работа

---

Для поиска спортивных личностей в баскетбольную команду университета необходимо автоматизировать анализ данных. 

Для этого в строке, подаваемой на вход и содержащей информацию о студенте (студентке), необходимо:

- проверить есть ли указание семейного положения в строке (эту информацию иногда забывают указать)

- проверить подходит ли студент(ка) для баскетбольной команды (рост должен быть не менее 180 для мужчин и 175 - для женщин, группа должна быть спорт или общая)

- вывести результат проверки в виде строки, содержащей информацию ФИО,рост,семейное положение (разделяются запятыми без пробелов)

---

**Ввод**: строка текста, содержащая информацию о студенте вида:


"ФИО,пол,группа,семейное положение(холост/женат/свободна/замужем),рост,физкульт.группа(общая/спорт/спец)"
ВНИМАНИЕ! Данные в строке разделяются запятыми. Может быть пропущена информация о семейном положении.

---

**Вывод**: 


- если студент подходит под требования баскетбольной команды и указано семейное положение (например "холост"):

"ФИО,181,холост"

- если студент(ка) подходит под требования баскетбольной команды, но семейное положение пропущено:

"ФИО,180,пропуск"

- если студентка не подходит под требования баскетбольной команды и указано семейное положение "замужем"


"ФИО,не подходит,замужем"

и т.д.

---

**Примеры**:

- для входной строки "Иванов И.Н.,муж,ПС-21,женат,187,общая" функция должна вернуть строку "Иванов И.Н.,187,женат"

- для входной строки "Магомедова А.М.,жен,ПС-32,,175,спорт" функция должна вернуть строку "Магомедова А.М.,175,пропуск"

---

**Подсказки**:

Для проверки вхождения подстроки ss в строку string используйте оператор in: 

```python
ss in string
```

Для поиска позиции i начала подстроки ss в строке string используйте метод find:

```python
i = string.find(ss)
```


````python
string = input()
# string = "Иванов И.Н.,муж,ПС-21,женат,187,общая"  # return "Иванов И.Н.,187,женат"
# string = "Магомедова А.М.,жен,ПС-32,,175,спорт" # return "Магомедова А.М.,175,пропуск"

lst = string.split(',')
di = {'FIO': None, 'Height_or_NotApproved': None, 'Status': None}

di['FIO'] = lst[0]

if (lst[5] == 'спорт' or lst[5] == 'общая') and \
      ( 
        (lst[1] == 'муж' and int(lst[4]) >= 180) or \
        (lst[1] == 'жен' and int(lst[4]) >= 175)
      ):
    di['Height_or_NotApproved'] = lst[4]
else:
    di['Height_or_NotApproved'] = 'не подходит'

if lst[3]:
    di['Status'] = lst[3]
else:
    di['Status'] = 'пропуск'

print(*di.values(), sep=',')
````
