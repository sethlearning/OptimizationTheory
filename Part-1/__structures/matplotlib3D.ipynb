```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)

ax.plot(xline, yline, zline, 'gray')
fig.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter(xdata, ydata, zdata, c=zdata, cmap='Greens')
ax.text2D(0.0, 0.9, ". 1.", transform=ax.transAxes)
fig.show()
```

```python
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)

print(Z.shape)
np.round(Z[0:5, 0:7], 2)
```

```python
from matplotlib import cm

fig, ax = plt.subplots(figsize=(8,6))
surf = ax.contourf(X, Y, Z, 30, cmap=plt.cm.Spectral)
ax.set_xlabel('x')
ax.set_ylabel('y')

ax.text(0.0, 0.0, '$(0, 0)$', color='black')
fig.suptitle('. 4.7.2.    $\sin {\sqrt {x^2 + y^2}}$', y=0)
fig.colorbar(surf, shrink=0.9, aspect=5)
fig.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

fig.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 100, cmap=cm.coolwarm)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.view_init(60, 30)
fig.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color='black')
ax.set_title('Wireframe')
fig.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='black')
fig.colorbar(surf, shrink=0.6, aspect=5, ax=ax)
ax.set_title('Surface')
fig.show()
```

```python
r = np.linspace(0, 6, 20)
theta = np.linspace(-0.9 * np.pi, 0.8 * np.pi, 40)
r, theta = np.meshgrid(r, theta)

X = r * np.sin(theta)
Y = r * np.cos(theta)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
fig.colorbar(surf, shrink=0.6, aspect=5, ax=ax)
ax.set_title('Surface')
fig.show()
```

```python
theta = 2 * np.pi * np.random.random(1000)
r = 6 * np.random.random(1000)
x = np.ravel(r * np.sin(theta))
y = np.ravel(r * np.cos(theta))

z = f(x, y)
print(z.shape)
print(z[0:10])
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(x, y, x, c=z, cmap='viridis', linewidth=0.5)
ax.set_title('Scatter')
fig.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_trisurf(x, y, z, cmap='viridis', edgecolor='none')
ax.set_title('Trisurf')
fig.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

zdirs = (None, 'x', 'y', 'z', (1, 1, 0), (1, 1, 1))
xs = (1, 4, 4, 9, 4, 0)
ys = (2, 5, 8, 10, 1, 0)
zs = (10, 3, 8, 5, 1, 0)

for zdir, x, y, z in zip(zdirs, xs, ys, zs):
    label = '(%d, %d, %d), dir=%s' % (x, y, z, zdir)
    ax.text(x, y, z, label, zdir)

ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_zlim(0, 10)

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

fig.show()
```
