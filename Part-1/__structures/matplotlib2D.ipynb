```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5, 5, 0.5)
y = x**2 + np.sin(x)
data = np.vstack([x, y])

print(np.round(data[:, :10], 3))

print(np.vsplit(data, 2))


plt.figure()
plt.scatter(x, y, marker='h', color='cornflowerblue', s=100, alpha=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5, 5, 0.5)
y = x**2 + np.sin(x)

plt.figure()
plt.plot(x,y, color='purple', marker='o', markersize=10, markerfacecolor='darkblue', markeredgecolor='lightblue', linestyle='dashed', linewidth=2.5)

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.xlabel('X', loc='left', fontdict=font2)
plt.ylabel('$y=x^2 + sin(x)$', loc='bottom', fontdict=font2)
plt.title("Title", fontdict=font1)

plt.grid(axis='both', color='green', linestyle='--', linewidth=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5, 5, 0.5)
y = x**2 + np.sin(x)

f = plt.figure(figsize=(5,4))

axes1 = f.add_axes([0, 0, 1, 1])
axes1.plot(x, y, 'r')

axes2 = f.add_axes([0.3, 0.5, 0.4, 0.4])
axes2.plot(y, x)

f.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5, 5, 0.5)
y = x**2 + np.sin(x)

f, axes = plt.subplots(2)

axes[0].plot(x, y, 'r')
axes[1].plot(y, x)

f.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5, 5, 0.5)
y = x**2 + np.sin(x)

f, axes = plt.subplots(1, 2, figsize=(10,6))

axes[0].plot(x, y, 'r')
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")
axes[0].set_title("First")

axes[1].plot(y, x)
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")
axes[1].set_title("Second")

f.show()
```
