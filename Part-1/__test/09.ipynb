# 3

```python
def f1(x):
  return pow(x**2, 1/3)
def f2(x):
  return np.log(1 + x**2)
```

```python
x = np.arange(-5, 5, 0.1)
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x, f1(x), 'r', f1(x), x, 'g')
axes[0].set_title('y=x^{2/3}')
axes[1].plot(x, f2(x), 'r', f2(x), x, 'g')
axes[1].set_title('y=log(1+x^2)')
fig.show()
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x, f1(x), 'r', f1(x), x, 'g')
axes[0].set_title('$y=x^{2/3}$')
axes[1].plot(x, f2(x), 'r', f2(x), x, 'g')
axes[1].set_title('$y=log(1+x^2)$')
fig.show()
```

```python
y1 = f1(x); y2 = f2(x)
plt.figure()
plt.plot(x,y1, y1, x)
plt.plot(x,y2, y2, x)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2)
axes[0].plot(x, f1(x), 'r', f1(x), x, 'g')
axes[1].plot(x, f2(x), 'r', f2(x), x, 'g')
plt.show()
```

```python
fig, axes = plt.subplots(2)
axes[0].plot(x, f1(x), 'r', f1(x), x, 'g')
axes[1].plot(x, f2(x), 'r', f2(x), x, 'g')
plt.show()
```
