1.4.2. Кластеризация методом K-means - массивы.

Тип работы: Индивидуальная работа

Пишем функцию `recalcXC(xtab)`, которая берет на вход **xtab** - таблицу (numpy array), содержащую в строках координаты точек - параметры объектов. 

Последний столбец таблицы содержит метку кластера (0 или 1).

Столбцы таблицы `xtab` соответствуют параметрам `x1`, `x2`, `label`.

Функция должна подсчитывать координаты центров кластеров с метками 0 и с метками 1.

На выходе функция должна возвращать `xctab` - таблицу (numpy array), содержащую в двух строках координаты точек - центров кластеров.

1-я строка - координаты XC0 (центр кластера с меткой 0), 2-я строка - XC1 (центр кластера с меткой 1). Столбцы таблицы  xctab соответствуют параметрам x1, x2.

Например, 

если `xtab = numpy.array( [ [0,1,1] , [1,0,0]])`, то `recalcXC(xtab) = numpy.array(  [ [1,0], [0,1] ] )`

(Здесь каждый кластер состоит всего из одной точки)

если `xtab = numpy.array( [ [0, 1, 1] , [1, 0, 0], [2, 0, 0], [0, 2, 1]] )`, то `recalcXC(xtab) = numpy.array(  [ [1.5, 0], [0, 1.5] ] )`

(Здесь каждый кластер состоит из двух точек)

```python
import numpy as np

# ----------------------- begin of user function --------------------------
# --- Эту функцию программирует обучающийся!!! ----
def recalcXC(xtab):
    '''
    xtab: таблицу (numpy.array), содержащую в строках координаты точек - параметры объектов. 
    Последний столбец таблицы содержит метку кластера (0 или 1). 
    Столбцы массива xtab соответствуют параметрам x1, x2,label. 
    Функция должна подсчитывать координаты центров кластеров как среднеарифметические координат 
    точек с метками 0 и с метками 1. На выходе функция должна возвращать xctab - таблицу (numpy.array), 
    содержащую в двух строках координаты точек - центров кластеров. 
    Столбцы массива  xctab соответствуют координатам x1, x2.
    1-я строка - координаты XC0 (центр кластера с меткой 0), 
    2-я строка - координаты XC1 (центр кластера с меткой 1). 
    '''
    xctab = np.zeros((2,2))

    for i in range(2):
        k = 0
        for x in xtab:
            if x[2] == i:
                xctab[i, 0] += x[0]
                xctab[i, 1] += x[1]
                k += 1
        if k > 1:
            xctab[i, 0] /= k
            xctab[i, 1] /= k

    return(xctab)

# ---------------------- end of user fun ---------------------------


# xtab = np.array( [ [0,1,1] , [1,0,0]])
# recalcXC(xtab) = numpy.array(  [ [1,0], [0,1] ] )

xtab = np.array( [ [0, 1, 1] , [1, 0, 0], [2, 0, 0], [0, 2, 1]] )
# recalcXC(xtab) = numpy.array(  [ [1.5, 0], [0, 1.5] ] )
recalcXC(xtab)
```
