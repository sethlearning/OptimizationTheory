### Program

```python
n = int(input('Введите количество сотрудников: '))
fiolist = []
for i in range(n):
    fiolist.append(input('Введите фамилию: '))

''' tests
# n = 6
# fiolist = ['Иванов', 'Петров', 'Сидоров', 'Петров', 'Иванов', 'Петров']
# fiolist = ['Сидоров', 'Иванов', 'Петров', 'Петров', 'Иванов', 'Петров']
# fiolist = ['Сидоров', 'ИВАНОВ', 'ПЕтров', 'Петров', 'Иванов', 'Петров']

n = 3
fiolist = ['Сидоров', 'Иванов', 'Петров']
'''

adict = {}
for fio in fiolist:
    fionorm = fio.lower().capitalize()
    if fionorm in adict:
        adict[fionorm] += 1
    else:
        adict[fionorm] = 1

singulars = []

for fio in adict:
    if adict[fio] == 1:
        singulars.append(fio)

for s in singulars:
    adict.pop(s)

if adict:
    print(adict)
else:
    print(0)
```
