Напишите функцию **board(num, size)**, создающую массив numpy с изображением квадратного чёрно-белого клетчатого поля.
В функцию поступают два целых числа — количество клеток n и размер клетки в пикселях s.
(в примере создаётся доска 8x8 клеток, размер каждой клетки — 50x50 пикселей).
Левая верхняя клетка должна быть чёрной.

Формат ввода

board(8, 50)

Формат вывода

**M** - numpy - массив размера (num * size, num * size, 3), при отображении которого с помощью plt.imshow(M) выводится изображение доски


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# создадим массив яркостей для черной полосы
black_band = np.zeros(shape=(20, 100, 3))

red_band = np.zeros(shape=(20, 100, 3))
red_band[:,:,0] = 1

green_band = np.zeros(shape=(20, 100, 3))
green_band[:,:,1] = 1

# создадим массив яркостей для белой полосы
white_band = np.ones(shape=(20, 100, 3))
```

```python
#im_array = np.concatenate([black_band, white_band, black_band, white_band], axis=0)
im_array = np.concatenate([red_band, white_band, green_band, black_band], axis=0)
print(im_array.shape)
plt.imshow(im_array);
```

### Program

```python
import numpy as np
import matplotlib.pyplot as plt

def board(n, size):
    brd = np.zeros(shape=(size*n, size*n, 3))
    for i in range(n):
        for j in range((i+1)%2, n, 2):
            brd[size*i : size*(i+1), size*j : size*(j+1)] = 1
    return brd
```

### Tests

```python
im = board(8, 10)
print(im.shape)
plt.imshow(im)
```
