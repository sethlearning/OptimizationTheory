1.3.3. Расстановка кораблей для морского боя

Тип работы:  Индивидуальная работа

Мы готовимся к игровым соревнованиям ботов. Игра в морской бой.

На данном этапе мы должны написать функцию формирования описания начального положения наших кораблей.

`start_boats_position(how_many_boats)`

**ВХОДНЫЕ ДАННЫЕ**:

`how_many_boats` - кортеж вида (n1, n2)

где n1, n2 - количество однопалубных и двухпалубных кораблей.

Пока мы будем тренироваться на малом поле 4х4. Поэтому будем расставлять не более 3-х однопалубных кораблей и не более одного двухпалубного:

`1 <= n1 <= 3; 0 <= n2 <= 1`

При этом мы немного ослабим правила морского боя и разрешим кораблям касаться углами друг друга. Но не стоять впритык сторонами.


Результатом работы функции должен являться список, который содержит расстановку кораблей n1 однопалубных кораблей и n2 двухпалубных (если n2=1).

При этом положение однопалубного корабля должно задаваться координатами клетки, в которой он расположен, так:

`[(letter, number)]`

где letter - это одна из букв a,b,c,d, а number - это целое число от 1 до 4.

Положение двухпалубного корабля должно задаваться так:

`[(letter1, number1), (letter2, number2)]`

При этом положение этих двух клеток должно быть смежным.


Пример. 

Пусть для входных параметров n1=2, n2=1 вы решили расположить корабли таким образом:

`start_boats_position`

Тогда результатом обращения к вашей функции 

`start_boats_position(how_many_boats = (2,1))`

должен быть такой список:

`[ [('a',4)], [('d',2)], [('b',1),('c',1)] ]`

Описания кораблей могут идти в любом порядке.

Необязательно, но будет лучше, если ваша функция будет рандомно выбирать позиции кораблей.

```python
''' --------- модуль слушателя курса -------- '''

# функция начальной расстановки кораблей 1-палубных и 2-х палубных
from numpy import random as rnd

def check_neighbors(t, bp):
    # returns True is there are no ships on the left, right, up or down
    # diagonal neighbors are allowed
    # `t` is proposed new position
    # `bp`` is flattened current boats_position

    for b in bp:
        # diagonal neighbor case
        if (abs(t[0] - b[0]) == 1) and (abs(t[1] - b[1]) == 1):
            continue
        if (abs(t[0] - b[0]) < 2) and (abs(t[1] - b[1]) < 2):
            return False

    return True

def flattenbp(bp):
    # flattens boats_position
    # i.e. [[(2, 3)], [(1, 2), (2, 2)]] -> [(2, 3), (1, 2), (2, 2)]
    res = []

    for b in bp:
        for t in b:
            res.append(t)

    return res

def replacecolswithletters(bp):
    # replace 1 with 'a',
    # 2 with 'b', etc.
    res = []
    cols = [*'_abcd']

    for i, b in enumerate(bp):
        res.append([])
        for t in b:
            res[i].append((cols[t[0]], t[1]))
    return res

def start_boats_position(how_many_boats):
    n1 = how_many_boats[0] # кол-во однопалубных
    n2 = how_many_boats[1] # кол-во двухпалубных

    boats_position = []

    if not (1 <= n1 <= 3 and 0 <= n2 <= 1):
        print("Количество 1-палубных кораблей: от 1 до 3\nКоличество 2-палубных кораблей: от 0 до 1")
        return

    if n2:
        r = rnd.randint(1, 5)
        c = rnd.randint(1, 5)
        lst = [(c, r)]

        while True:
            direction = rnd.randint(1, 5)
            # 1 - up, 2 - right, 3 - down, 4 - left
            if direction == 1 and r != 1:
                r -= 1
                break
            elif direction == 2 and c != 4:
                c += 1
                break
            elif direction == 3 and r != 4:
                r += 1
                break
            elif direction == 4 and c != 1:
                c -= 1
                break
        lst.append((c, r))
        boats_position.append(lst)

    for _ in range(n1):
        flattenedbp = flattenbp(boats_position)
        while True:
            r = rnd.randint(1, 5)
            c = rnd.randint(1, 5)
            if check_neighbors((c, r), flattenedbp):
                boats_position.append([(c, r)])
                break

    boats_position = replacecolswithletters(boats_position)
    return boats_position

# start_boats_position(how_many_boats=(3,1))
```
