2.4.1. Оцените количество изюма в булочках

**Тип работы**:  Индивидуальная работа

Закон Пуассона можно использовать для того, чтобы описать закон распределения вероятности того, что в булочке окажется k изюма, если известно среднее количество изюма в булочках λ по формуле:

pλ(k)=λk⋅e−λk!

**ЗАДАНИЕ**. Запрограммируйте функцию check_prob(l, k) - вероятность того, что в булочке изюма окажется меньше или равно числа k.

--------------------------

**ВХОД**:

`l` - среднее число изюма в булочке, тип float;

`k` - целое неотрицательное число;

--------------------------

**ВЫХОД**:

`p` - вероятность того, что в булочке изюма окажется меньше или равно числа k, округленная до 5-го знака после запятой

--------------------------

**ПРИМЕР**:

При обращении check_prob(l=2, k=2) выход должен быть равен 0.67668

При обращении check_prob(l=5, k=2) выход должен быть равен 0.12465

При обращении check_prob(l=10, k=5) выход должен быть равен 0.06709

Проверка производится с точностью до 5-го знака после запятой

```python
import numpy as np
import math
def check_prob(l, k):
    """
    ВХОД:
    l - среднее число изюма в булочке, тип float;
    k - целое неотрицательное число;
    
    ВЫХОД:
    p - вероятность того, что в булочке изюма окажется меньше или равно числа k, округленная до 5-го знака после запятой
    """
    p = 0
    for i in range(k+1):
        p += l**i*np.exp(-l)/math.factorial(i)
    return np.round(p, 5)


# check_prob(2, 2)  # 0.67668

# check_prob(5, 2)  # 0.12465

# check_prob(10, 5)  # 0.06709
```
