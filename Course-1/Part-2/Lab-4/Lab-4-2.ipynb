2.4.2. Удастся ли провести занятие в онлайн-среде?

**Тип работы**:  Индивидуальная работа

С начала периода пандемии (с марта 2020 года) все обучение ушло в онлайн. И Университету пришлось организовать проведение лекций с использованием вебинарной системы BigBlueButton, реализованного в рамках портала MOODLE. Было поднято несколько серверов и факультеты развели по разным "кнопкам"-серверам. Тем не менее сервер перестаёт справляться с обслуживанием вебинаров, если количество подключенных пользователей оказывалось более 800 (примерно 16 лекционных потоков). Среднее количество подключенных пользователей составляет 500 человек на сервер (10 лекционных потоков). 

Оценить вероятность того, что вебинарный сервер не справится с нагрузкой (т.е. количество пользователей превысит 800 человек), если считаем, что количество подключенных пользователей к одному серверу подчиняется нормальному закону распределения со средним m и СКО σ.

**ЗАДАНИЕ**: Написать функцию check_prob(m, sigma)

**ВХОД**: `m`, `sigma` - среднее и СКО нормального распределения

**ВЫХОД**: `p(X > 800)` , округленное до 5-го знака после запятой

**ПРИМЕР**:

При вычислении check_prob(m=500, sigma=100) выход должен быть 0.00137 с точностью до 5-го знака после запятой

При вычислении check_prob(m=600, sigma=100) выход должен быть 0.02302 с точностью до 5-го знака после запятой

При вычислении check_prob(m=600, sigma=200) выход должен быть 0.16060 с точностью до 5-го знака после запятой

```python
import numpy as np
import scipy.stats as stat

def check_prob(m, sigma):
    """
    ВХОД: m, sigma - среднее и СКО нормального распределения

    ВЫХОД: p(X > 800) , округленное до 5-го знака после запятой
    """
    x = 800
    cdf = stat.norm.cdf(x, loc=m, scale=sigma)
    p = 1 - cdf
    return np.round(p, 5)


# check_prob(m=500, sigma=100)
# check_prob(m=600, sigma=100)
# check_prob(m=600, sigma=200)
```
