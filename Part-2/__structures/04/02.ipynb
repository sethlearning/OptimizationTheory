```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
x = np.arange(1, 5, dtype=int)
nn = np.array([35, 5, 50, 10])

p = nn / sum(nn)

print(x, p)
```

```python
plt.scatter(x, p, s=50)
for i in range(len(x)):
    plt.plot([x[i], x[i]], [0, p[i]], c='grey')
plt.suptitle('Полигон относительных частот для Д.С.В., принимающей значения 1, 2, 3, 4', y=0)
plt.show()
```
