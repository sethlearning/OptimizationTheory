```python
import numpy as np
import scipy.stats as stat
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
x = np.linspace(-3, 3, 100)
mu1 = 1.0
sigma1 = 0.5
norm_rv1 = stat.norm(loc=mu1, scale = sigma1)
mu2 = 0.0
sigma2 = 1.0
norm_rv2 = stat.norm(loc=mu2, scale = sigma2)

pdf1 = norm_rv1.pdf(x)
pdf2 = norm_rv2.pdf(x)

plt.plot(x, pdf1, label=f'm={mu1}, $\sigma$={sigma1}')
plt.plot(x, pdf2, label=f'm={mu2}, $\sigma$={sigma2}')
plt.ylabel('$\phi(x, m, \sigma)$')
plt.xlabel('x')
plt.legend()
plt.suptitle('Плотности нормального распределения с параметрами $m, \sigma$', y=0)
plt.show()
```

```python
mu = 5.0
sigma = 2
norm_rv = stat.norm(loc=mu, scale = sigma)

xvals = norm_rv.rvs(size=100)
print(xvals[:10])

mu2 = xvals.mean()
S2 = stat.moment(xvals, moment=2)
sigma2 = np.sqrt(S2)
print(mu2, S2, sigma2)

norm_rv2 = stat.norm(loc=mu2, scale = sigma2)
x = np.linspace(-5, 15, 100)

pdf = norm_rv.pdf(x)
pdf2 = norm_rv2.pdf(x)

plt.plot(x, pdf, label=f'Исходная с m={mu1}, $\sigma$={sigma1}')
plt.plot(x, pdf2, label=f'Оцененная с $\hat m$={round(mu2, 2)}, $\hat \sigma$={round(sigma2, 2)}')
plt.ylabel('$f_X(x)$')
plt.xlabel('x')
plt.legend()
plt.suptitle('Исходная и оцененная плотности нормального распределения', y=0)
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
mu1 = 1.0
sigma1 = 0.5
norm_rv1 = stat.norm(loc=mu1, scale = sigma1)
mu2 = 0.0
sigma2 = 1.0
norm_rv2 = stat.norm(loc=mu2, scale = sigma2)

сdf1 = norm_rv1.cdf(x)
сdf2 = norm_rv2.cdf(x)

plt.plot(x, сdf1, label=f'm={mu1}, $\sigma$={sigma1}')
plt.plot(x, сdf2, label=f'm={mu2}, $\sigma$={sigma2}')
plt.ylabel('$\phi(x, m, \sigma)$')
plt.xlabel('x')
plt.legend()
plt.suptitle('Функции нормального распределения с параметрами $m, \sigma$', y=0)
plt.show()
```
