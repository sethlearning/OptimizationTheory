```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
x = np.linspace(-3, 3, 101, endpoint=True)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16, 4))

ax1.plot(x, x**3)
ax1.set_ylabel('$y = x^3$')
ax1.set_title('(a) Непрерывная функция')

ax2.plot(x[x<0], np.sign(x[x<0]), 'b')
ax2.plot(x[x>0], np.sign(x[x>0]), 'b')
ax2.scatter([0.0], [0.0], s=50, c='b')
ax2.set_ylabel('$y = sign(x)$')
ax2.set_title('(б) Разрыв первого рода')

ax3.plot(x[x<0], 1/x[x<0], 'g')
ax3.plot(x[x>0], 1/x[x>0], 'g')
ax3.set_ylabel('$y = 1/x$')
ax3.set_title('(в) Разрыв второго рода')

fig.suptitle('Примеры функций', y=0)
```

```python
x = np.linspace(-3, 3, 100, endpoint=True)
yfun = lambda x: x**3 - 5*x

xmin, xmax = np.sqrt(5/3), -np.sqrt(5/3)

plt.figure(figsize=(6, 4))
plt.plot(x[x < xmax], yfun(x[x < xmax]), 'r')
plt.plot(x[x > xmax], yfun(x[x > xmax]), 'g')
plt.plot(x[x > xmin], yfun(x[x > xmin]), 'r')
plt.scatter([xmax, xmin], [yfun(xmax), yfun(xmin)], s=50, c='b')
plt.suptitle('Интервалы возрастания-убывания функции', y=0)
plt.show()
```

```python
x = np.linspace(-2, 4, 100, endpoint=True)
yfun = lambda x: (x-1)**3 - 5*x
x_per = 1.0

plt.figure(figsize=(6, 4))
plt.plot(x[x < x_per], yfun(x[x < x_per]), c='r')
plt.plot(x[x > x_per], yfun(x[x > x_per]), c='g')
plt.scatter(x_per, yfun(x_per), s=50, c='b')
plt.suptitle('Интервалы выпуклости вверх и вниз', y=0)
plt.show()
```

```python
x = np.linspace(-2, 2, 101, endpoint=True)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))
y1 = x**3
ax1.plot(x, y1, 'g')
ax1.plot(y1, x, 'r')
ax1.plot(x, x, color='grey', linestyle='dashed')
ax1.set_ylabel('$F(x)=x^3; F^{-1}(x)=\sqrt[3]{x} $')
ax1.set_title('Обратимая функция')

y2 = x**2
ax2.plot(x[x<0], y2[x<0], 'b')
ax2.plot(x[x>0], y2[x>0], 'g')
ax2.plot(x[x>0], x[x>0]**(1/2), 'r')
ax2.scatter([0.0], [0.0], s=50, c='grey')
ax2.plot(x, x, color='grey', linestyle='dashed')
ax2.set_ylabel('$F(x)=x^2; F^{-1}(x)=\sqrt{x}, x > 0 $')
ax2.set_title('Необратимая функция')
fig.suptitle('Примеры обратимой и необратимой функций', y=0)
plt.show()


```
