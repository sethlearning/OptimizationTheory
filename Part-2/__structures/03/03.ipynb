```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
x = np.linspace(0, 2, 100, endpoint=True)
yfun = lambda x: x**2
x0, x1 = 1.0, 1.5

plt.figure(figsize=(6,6))
plt.plot(x, yfun(x), 'g')
plt.plot([x0, x1], [yfun(x0), yfun(x1)], 'r')
plt.plot([x0, x1, x1], [yfun(x0), yfun(x0), yfun(x1)], 'grey')
plt.text(1.25, 0.7, '$\Delta x$')
plt.text(1.6, 1.5, '$\Delta y$')
plt.text(0.75, 1.0, '$x_0, y_0$')
plt.text(1.25, 2.4, '$x_1, y_1$')
plt.scatter([x0, x1, x1], [yfun(x0), yfun(x0), yfun(x1)], s=50, c='b')
plt.suptitle('Оценка скорости изменения функции в точке $x_0$', y=0)
plt.show()
```

```python
x = np.linspace(0, 2, 100, endpoint=True)
yfun = lambda x: x**3

x_i = np.arange(1.0, 2.1, 0.125)
print(x_i)
x_i[list(range(1, len(x_i), 2))] = x_i[0]
print(x_i)
x_i = x_i[1:]
print(x_i)

plt.figure(figsize=(6,6))
plt.plot(x, yfun(x), 'g')
plt.plot(x_i, yfun(x_i), 'r')
plt.scatter(x_i, yfun(x_i), s=50, c='b')
plt.text(1.05, 1.0, '$x_0, y_0$')
plt.text(1.25, 1.5, '$x_4, y_4$')
plt.text(1.5, 3, '$x_3, y_3$')
plt.text(1.8, 5, '$x_2, y_2$')
plt.text(1.75, 8, '$x_1, y_1$')
plt.suptitle('Оценка скорости изменения функции в точке $x_0$', y=0)
plt.show()
```

```python
x = np.linspace(-2, 2, 100, endpoint=True)
yfun = lambda x: x**2
x0, x1, x2 = -1.5, 0.0, 1.0
k0, k1, k2 = -3, 0.0, 2.0
dx = np.linspace(-0.9, 0.9, 51, endpoint=True)

plt.figure(figsize=(5, 6))
plt.plot(x, yfun(x), 'g')
plt.plot(x0 + dx, yfun(x0) + k0 * dx, 'r')
plt.plot(x1 + dx, yfun(x1) + k1 * dx, 'r')
plt.plot(x2 + dx, yfun(x2) + k2 * dx, 'r')
plt.scatter([x0, x1, x2], [yfun(x0), yfun(x1), yfun(x2)], s=50, c='b')
plt.suptitle('Касательные к функции $x^2$', y=0)

plt.show()
```

```python
x = np.linspace(-2, 2, 100, endpoint=True)
yfun = lambda x: x**3 - 2 * x
y1fun = lambda x: 3 * x**2 - 2
y2fun = lambda x: 6 * x

x0, x1, x2 = -np.sqrt(2/3), 0.0, np.sqrt(2/3)

plt.figure(figsize=(5,6))
plt.plot(x, yfun(x), 'r', label='$f(x)=x^3-2x$')
plt.plot(x, y1fun(x), 'b', label="$f'(x)$")
plt.plot(x, y2fun(x), 'g', label="$f''(x)$")
plt.scatter([x0, x1, x2], [y1fun(x0), yfun(x1), y1fun(x2)], s=50, c='b')
plt.suptitle('Производные к функции $f(x)=x^3-2x$', y=0)

plt.legend()
plt.show()
```

```python
d = np.linspace(0, 5, 100, endpoint=True)
tfun = lambda d: 0.5*(d-3)**3 + 2*d + 13.5

d1, d2 = 0.0, 3.0
k1, k2 = 15.5, 2
dd = np.linspace(-0.1, 1.5, 51, endpoint=True)

plt.figure(figsize=(6,6))
plt.plot(tfun(d), d, 'g')
plt.scatter([tfun(d1), tfun(d2)], [d1, d2], s=50, c='b')
plt.plot(tfun(d1)+k1*dd, d1+dd, 'r')
plt.plot(tfun(d2)+k2*dd, d2+dd, 'r')
plt.xlabel('t, мин')
plt.ylabel('d, км')
plt.suptitle('Зависимость пройденного расстояния от времени', y=0)
plt.show()
```
