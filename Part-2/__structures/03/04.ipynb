```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
x = np.linspace(-1, 1, 100, endpoint=True)
yfun = lambda x: np.exp(x) - 1

plt.figure(figsize=(6,6))
plt.plot(x, yfun(x), 'g', label='$y=e^x-1$')
plt.plot(x, x, 'grey', label='$y=x$')
plt.plot([-0.2, 0.2], [-0.2, 0.2], 'r')
plt.scatter([0], [0], s=50, c='b')
plt.suptitle('Графики функций $e^x-1, \, x$ вблизи нуля')

plt.legend()
plt.show()
```
