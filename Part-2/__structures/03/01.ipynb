```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
xmin , xmax = -3.4, 3.4
x1 = np.linspace(xmin, xmax, 500)
y1 = np.sin(x1)
z1 = (x1 - 1.0)**2 + (y1 + 1)**4
dots_1 = np.vstack([x1, y1, z1])
noise = 0.2 * np.random.randn(*dots_1.shape)
dots_1 += noise

dots_2 = np.random.randn(*dots_1.shape)
dots_all = np.hstack([dots_1, dots_2])
print(dots_1.shape, dots_all.shape)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=[9, 6])
# ax = fig.gca(projection='3d')
ax = plt.axes(projection='3d')
# ax = plt.subplot(projection='3d')
ax.view_init(30, -60)

ax.scatter(dots_1[0], dots_1[1], dots_1[2], s=50, marker='o')
ax.scatter(dots_2[0], dots_2[1], dots_2[2], s=50, marker='+')

plt.suptitle('Рис. 1. Два семейства точек в трехмерном пространстве', y=0)
plt.show()
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, perplexity=50)
tsne.fit(dots_all.T)
Z = tsne.embedding_
print(Z.shape)
```

```python
plt.suptitle('Рис. 2. Два семейства точек, спроецированные в двумерное пространство', y=0)
plt.scatter(Z[:Z.shape[0] // 2, 0], Z[:Z.shape[0] // 2, 1], c='r')
plt.scatter(Z[Z.shape[0] // 2:, 0], Z[Z.shape[0] // 2:, 1], c='b')
plt.show()
```
