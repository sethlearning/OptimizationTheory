```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

from mpl_toolkits import mplot3d
```

```python
y_net = np.array([[3, 12, 8], [4, 8, 9], [5, 9, 12]])
y_goal = np.array([5, 10, 10])

ax = plt.axes(projection='3d')
ax.scatter(y_net[:, 0], y_net[:, 1], y_net[:, 2], s=100, c='blue')
ax.scatter(y_goal[0], y_goal[1], y_goal[2], s=50, c='red')
ax.view_init(30, 30)
ax.plot(y_net[:, 0], y_net[:, 1], y_net[:, 2])
```
