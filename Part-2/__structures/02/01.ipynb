```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

from mpl_toolkits import mplot3d
```

```python
eps = np.array([[3, -1, 1], [2, -2, 1],[1.5, -1.5, -1.5]])
ax = plt.axes(projection='3d')
ax.scatter(eps[:, 0], eps[:, 1], eps[:, 2], s=300, c='blue')
ax.scatter(0, 0, 0, s=50, c='red')
ax.view_init(30, 30)
ax.plot(eps[:, 0], eps[:, 1], eps[:, 2])
plt.show()
```

```python
def xy_norm1(fi):
    R = 1/(np.abs(np.cos(fi)) + np.abs(np.sin(fi)))
    return np.array([R*np.cos(fi), R*np.sin(fi)])

def xy_norm2(fi):
    return np.array([np.cos(fi), np.sin(fi)])

def xy_norm_inf(fi):
    R = np.min(np.vstack([1.0/np.abs(np.cos(fi)), 1.0/np.abs(np.sin(fi))]), axis=0)
    return np.array([R * np.cos(fi), R * np.sin(fi)])

fi = np.arange(0.03, 2*np.pi, np.pi/180)
x1 = np.vstack([1.0/np.abs(np.cos(fi)), 1.0/np.abs(np.sin(fi))])

xy_1 = xy_norm1(fi)
xy_2 = xy_norm2(fi)
xy_inf = xy_norm_inf(fi)

plt.figure(figsize=(6, 6))
plt.plot(xy_1[0], xy_1[1], 'r', label='$||x||_1 = 1$')
plt.plot(xy_2[0], xy_2[1], 'b', label='$||x||_2 = 1$')
plt.plot(xy_inf[0], xy_inf[1], 'g', label='$||x||_{\infty} = 1$')

plt.scatter(x = [1.0, 0.5, 1.0], y = [0.0, -0.5, -1.0], c='orange', s = 75)
plt.legend()
plt.show()
```

```python
x = (1.0, -1.0)
print(f'||(1.0, -1.0)||_1 = {np.linalg.norm(x, ord=1)}')
print(f'||(1.0, -1.0)||_2 = {np.linalg.norm(x, ord=2)}')
print(f'||(1.0, -1.0)||_inf = {np.linalg.norm(x, ord=np.inf)}')
```
