```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

from mpl_toolkits import mplot3d
```

```python
def f(x, eps=2.0, random_state=20):
    np.random.seed(random_state)
    y = 2 - 5 * x + x**2 + (np.random.random(size = len(x)) - 0.5)*eps
    return y

x_list = np.arange(0.0, 7.0, 1.0)
y_list = f(x_list)

x1_arr = np.arange(0.0, 6.1, 0.2)

plt.hlines(0, x_list.min(), x_list.max())
plt.vlines(0, y_list.min(), y_list.max())
plt.plot(x1_arr, f(x1_arr, eps=0), 'g--')
plt.scatter(x_list, y_list, s=100, marker='X')
plt.show()
```
