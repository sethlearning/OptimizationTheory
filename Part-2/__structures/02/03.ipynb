```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

from mpl_toolkits import mplot3d
```

```python
x = np.array([1, 2])
y = np.array([4, -2])
zero = np.array([0, 0])

plt.figure(figsize=(5, 5))
plt.scatter([x[0], y[0], zero[0]], [x[1], y[1], zero[1]])
plt.plot([x[0], zero[0]], [x[1], zero[1]], c='grey')
plt.plot([y[0], zero[0]], [y[1], zero[1]], c='grey')
plt.xlim((-1, 5))
plt.ylim((-3, 3))
plt.show()

print(f'(x, y) = {x @ y}')
```

```python

xyz = np.array([[0, 0, 0], [1, 2, 3], [3, 2, 1]])

ax = plt.axes(projection = '3d')
ax.scatter(xyz[:, 0], xyz[:, 1], xyz[:, 2], s=200, c='blue')

ax.plot(xyz[0:2, 0], xyz[0:2, 1], xyz[0:2, 2], c='grey')
ax.plot(xyz[0:3:2, 0], xyz[0:3:2, 1], xyz[0:3:2, 2], c='grey')
ax.view_init(30, 30)
plt.show()

print(f'(x, y) = {xyz[1] @ xyz[2]}')
phi = np.arccos(xyz[1] @ xyz[2] / np.linalg.norm(xyz[1]) / np.linalg.norm(xyz[2])) * 180 / np.pi
print(f'phi(x, y) = {round(phi, 2)} gr')
```

```python
phi = np.arccos(np.sqrt(5) / 3) * 180 / np.pi
print(f'phi(x, y) = {round(phi, 2)} gr')
```

```python
xyz = np.array([[0, 0, 0], [1, 2, 3], [3, 2, 1]])
print(np.linalg.norm(xyz[1]), np.linalg.norm(xyz[2]))
print(np.linalg.norm(xyz[1], ord=1), np.linalg.norm(xyz[2], ord=1))
print(np.linalg.norm(xyz[1], ord=2), np.linalg.norm(xyz[2], ord=2))
print(np.linalg.norm(xyz[1], ord=np.Inf), np.linalg.norm(xyz[2], ord=np.Inf))

print(xyz[1, 0]**2 + xyz[1, 1]**2 + xyz[1, 2]**2)
print(np.sqrt(xyz[1, 0]**2 + xyz[1, 1]**2 + xyz[1, 2]**2))
```
